# Unit Conversion

```python
df['Height'].dtype
```

```python
df['Height'].unique()
```

```python
def convert_height(height):
    if 'cm' in height:
        return int(height.strip('cm'))
    else:
        feet, inches = height.split('\'')
        total_inches = int(feet)*12 + int(inches.strip('\"'))
        return round(total_inches*2.54)

df['Height'] = df['Height'].apply(convert_height)
df['Height'].unique()
```

```python
df = df.rename(columns = {'Height':'Height (cm)'})
```

```python
df['Weight'].dtype
```

```python
df['Weight'].unique()
```

```python
def convert_weight(weight):
    if 'kg' in weight:
        return int(weight.strip('kg'))
    else:
        pounds = int(weight.strip('lbs'))
        return round(pounds/2.205)
    
df['Weight'] = df['Weight'].apply(convert_weight)
df['Weight'].unique()
```

```python
df = df.rename(columns = {'Weight':'Weight (kg)'})
```

```python
df[['Height (cm)', 'Weight (kg)']]
```

# Cleaning and Transforming Value, Wage and Release Clause Columns 

```python
df[['Value', 'Wage', 'Release Clause']]
```

```python
df.dtypes[['Value', 'Wage', 'Release Clause']]
```

```python
def ConvertUnit(unit:str):
    if '€' in unit:
        unit = unit.replace('€', '')
    if 'M' in unit:
        return int(float(unit.replace('M', ''))*1000000)
    elif 'K' in unit:
        return int(float(unit.replace('K', ''))*1000)

df['Value'] = df['Value'].map(ConvertUnit)
df['Wage'] = df['Wage'].map(ConvertUnit)
df['Release Clause'] = df['Release Clause'].map(ConvertUnit)

df[['Value', 'Wage', 'Release Clause']]
```
